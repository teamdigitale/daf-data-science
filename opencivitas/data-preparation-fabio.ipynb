## Merge dei Dati

```python
df_spese_comuni_merge = pd.merge(
    df_spesa_comuni, 
    df_comuni, 
    how='left',
    right_on='Codice Nazionale', 
    left_on='COMUNE_CAT_COD')
df_spese_comuni_merge.shape
```

```python
df_spese_comuni_merge.head()
```

```python
df_spese_comuni_merge.ANNO.value_counts()
```

## Merge Dati con [DatiIRPEF](http://localhost:8888/notebooks/mise_aiuti/DAF_MEF_DatiIRPEF.ipynb)

```python
df_irpef_comuni = pd.read_csv('../mise_aiuti/data/redditi_irpef.csv')
df_irpef_comuni = df_irpef_comuni.drop(df_irpef_comuni.columns[0], axis=1)
```

I dati relativi all'irpef coprono il periodo 2009-2015. Mentre quelli sulle spese di open sose vanno dal 2010 al 2013

```python
df_irpef_comuni['Anno di imposta'].value_counts()
```

```python
df_spese_comuni_merge['ANNO'].value_counts()
```

```python
print('shape irpef comuni ',df_irpef_comuni.shape)
print('shape spese comuni', df_spese_comuni_merge.shape)
```

We restrict the analysis on data in the period 2010-2013. Thus, we join the data from the two datasets based on cod_comune and year

```python
df_spese_comuni_irpef = pd.merge(
    df_spese_comuni_merge, 
    df_irpef_comuni, 
    how='left',
    left_on=['COMUNE_CAT_COD', 'ANNO'], 
    right_on=['Codice catastale', 'Anno di imposta'])
df_spese_comuni_irpef.shape
```

```python
df_spese_comuni_irpef['ANNO'].value_counts()
```

```python
df_spese_comuni_irpef.head()
```

```python
df_final = df_spese_comuni_irpef
```

```python
is_null = df_final.isnull().all() / len(df_final)
is_null[is_null > 0.0]
```
