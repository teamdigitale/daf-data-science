```python
to_drop = [
    'Attesa VCT Fabbricati', 
    'Attesa VCT Territorio',
    'Data Costituzione',
    'Denominazione Estera',
]


to_merge = [
    [
        "Reddito da lavoro autonomo (comprensivo dei valori nulli) - Ammontare", 
        "Reddito da lavoro autonomo (compresi nulli) - Ammontare"
    ],
    [
        "Reddito da lavoro autonomo (comprensivo dei valori nulli) - Frequenza",
        "Reddito da lavoro autonomo (compresi nulli) - Frequenza"
    ],
    [
        "Reddito da partecipazione  (comprensivo dei valori nulli) - Ammontare", 
        "Reddito da partecipazione (compresi nulli) - Ammontare"
    ],
    [
        "Reddito da partecipazione  (comprensivo dei valori nulli) - Frequenza", 
        "Reddito da partecipazione  (compresi nulli) - Frequenza"
    ],
    [
        "Reddito di spettanza dell'imprenditore in contabilita' ordinaria  (comprensivo dei valori nulli) - Ammontare",
        "Reddito spettanza imprenditore ordinaria  (compresi nulli) - Ammontare"
    ],
    [
        "Reddito di spettanza dell'imprenditore in contabilita' ordinaria  (comprensivo dei valori nulli) - Frequenza",
        "Reddito spettanza imprenditore ordinaria  (compresi nulli) - Frequenza"
    ],
    [
        "Reddito spettanza imprenditore semplificata (compresi nulli) - Ammontare",
        "Reddito di spettanza dell'imprenditore in contabilita' semplificata (comprensivo dei valori nulli) - Ammontare"
    ],
    [
        "Reddito spettanza imprenditore semplificata (compresi nulli) - Frequenza",
        "Reddito di spettanza dell'imprenditore in contabilita' semplificata (comprensivo dei valori nulli) - Frequenza"
    ],
    [
        'Sigla Provincia_x', 
        'Sigla Provincia_y'
    ],
    [
        'Codice Istat_x',
        'Codice Istat_y'
    ],
    [
        'Reddito complessivo minore di zero euro - Frequenza',
        'Reddito complessivo minore o uguale a zero euro - Frequenza',
    ],
    [
        'Reddito complessivo minore di zero euro - Ammontare',
        'Reddito complessivo minore o uguale a zero euro - Ammontare',
    ]
    
]

to_merge_name = [
    'Reddito lavoro autonomo Ammontare',
    'Reddito lavoro autonomo Frequenza',
    'Reddito da partecipazione Ammontare',
    'Reddito da partecipazione Frequenza',    
    'Reddito imprenditore ordinario Ammontare',
    'Reddito imprenditore ordinario Frequenza',
    'Reddito imprenditore semplificata Ammontare',
    'Reddito imprenditore semplificata Frequenza',
    'Provincia',
    'Cod Istat',
    'Reddito minore o uguale a zero Frequenza',
    'Reddito minore o uguale a zero Ammontare',
]
 
```

```python
import math
def merge_columns_data(row):
    a, b = row.values
    if type(a) is str:
        return a
    elif type(b) is str:
        return b
    elif math.isnan(a) and math.isnan(b):
        return a
    elif math.isnan(a):
        return b
    else:
        return a
```

```python
for i in range(len(to_merge)):
    cols = to_merge[i]
    name = to_merge_name[i]
    df_final[name] = df_final[cols].apply(merge_columns_data,axis=1)
```

```python
merged_to_drop = [val for sublist in to_merge for val in sublist] + to_drop
merged_to_drop
```

```python
df_final.shape
```

```python
df_final = df_final.drop(merged_to_drop, axis=1)
```

## Rename Colonne Costi Servizi

```python
df= pd.read_csv('./data/opencivitas_spesa_storica_dati_irpef.csv')
```

```python
costi_columns = [v for v in df.columns if v.startswith('SPESA')]
costi_columns_rename = [
    'riscossione tributi',
    'ufficio tecnico',
    'anagrafe, stato civile',
    'altri servizi',
    'polizia locale',
    'istruzione',
    "viabilita'",
    "trasporti",
    "territorio",
    "ambiente",
    "sociale",
    "asili nido",
    "spesa totale"
]

costi_columns_dict = dict(zip(costi_columns, costi_columns_rename))

tot_costi_columns = [v for v in df.columns if v.startswith('TOT_SPESA')]
tot_costi_columns_rename = ['tot ' + v for v in costi_columns_rename]

tot_costi_columns_dict = dict(zip(tot_costi_columns, tot_costi_columns_rename))
```

```python
tot_costi_columns_dict
```

```python
columns_renamed = []
for v in df.columns:
    if v in costi_columns_dict:
        columns_renamed.append(costi_columns_dict[v])
    elif v in tot_costi_columns_dict:
        columns_renamed.append(tot_costi_columns_dict[v])
    else:
        columns_renamed.append(v)
```

```python
df.columns = columns_renamed
```
